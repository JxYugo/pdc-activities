```python
import time
import threading
```

```python
dishes = ["Burger","Raw Liver","Goyslop","Soylent","Raw Milk"]

def prep_dish(dish):
  print(f"Prepping dish: {dish}")
  time.sleep(3)

def cook_dish(dish):
  print(f"Cooking Dish: {dish}")
  time.sleep(3)

def garnish_dish(dish):
  print(f"Garnish Dish: : {dish}")
  time.sleep(3)

def serve_dish(dish):
  print(f"Serving Dish: : {dish}")
  time.sleep(3)
```

```python
# Sequential

start_time =  time.time()
for dish in dishes:
  prep_dish(dish)
  cook_dish(dish)
  garnish_dish(dish)
  serve_dish(dish)
end_time =  time.time()

execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```

```python
# Parallel

def parallel_dish(dish):
    prep_thread = threading.Thread(target=prep_dish, args=(dish,))
    cook_thread = threading.Thread(target=cook_dish, args=(dish,))
    garnish_thread = threading.Thread(target=garnish_dish, args=(dish,))
    serve_thread = threading.Thread(target=serve_dish, args=(dish,))

    prep_thread.start()
    cook_thread.start()
    garnish_thread.start()
    serve_thread.start()

    prep_thread.join()
    cook_thread.join()
    garnish_thread.join()
    serve_thread.join()
```

```python
start_time = time.time()

threads = []
for dish in dishes:
    t = threading.Thread(target=parallel_dish, args=(dish,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

end_time = time.time()

execution_time = end_time - start_time
print(f"\nExecution time (Parallel): {execution_time:.2f} seconds")
```
